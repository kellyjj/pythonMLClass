```python
dfs = [allsites_df,fandango_df]
combo_df = pd.concat(dfs,axis=1,join="inner")
combo_df
```

```python
combo_df['IMDB_Norm'] = np.round(combo_df['IMDB']/2,1)
```

```python
combo_df['MetaCritic_Norm'] = np.round(combo_df['Metacritic']/20,1)
combo_df['MetaCritic_U_Norm'] = np.round(combo_df['Metacritic_User']/2,1)
```

```python
combo_df['RT_Norm'] = np.round(combo_df['RottenTomatoes']/20,1)
combo_df['RT_U_Norm'] = np.round(combo_df['RottenTomatoes_User']/20,1)
```

```python
comboNorm_df = combo_df[['STARS','RATING','IMDB_Norm','MetaCritic_Norm','MetaCritic_U_Norm','RT_Norm','RT_U_Norm']]
#combo_df.head()
```

```python
comboNorm_df
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.kdeplot(data=comboNorm_df,x='STARS',fill=True,label="Stars")
sns.kdeplot(data=comboNorm_df,x='RATING',fill=True,label="Rating")
sns.kdeplot(data=comboNorm_df,x='IMDB_Norm',fill=True,label="IMDB_Norm")
sns.kdeplot(data=comboNorm_df,x='MetaCritic_Norm',fill=True,label="MetaCritic_Norm")
sns.kdeplot(data=comboNorm_df,x='RT_Norm',fill=True,label="RT_Norm")
sns.kdeplot(data=comboNorm_df,x='MetaCritic_U_Norm',fill=True,label="MetaCritic_U_Norm")
sns.kdeplot(data=comboNorm_df,x='RT_U_Norm',fill=True,label="RT_U_Norm")
plt.legend(loc=(1.05,0.5))
```
