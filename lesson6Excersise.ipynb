```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m=np.linspace(0,10,11)
c = 3 * 10**8
E=m*c**2
```

```python
m
```

```python
E
```

```python
relfig = plt.figure()
relaxes = relfig.add_axes([0,0,1,1])
relaxes.set_ylabel("Energy")
relaxes.set_xlabel("Mass")
#relaxes.set_ylim(0,8)
relaxes.set_xlim(0,10)
relaxes.plot(m,E,color='gold',lw=2)
relfig.show()
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)
plt.show()
```
