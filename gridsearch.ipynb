```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge


df = pd.read_csv('/home/kellyj/Programing/PythonProjects/ML_Udemy/NoteBooks/UNZIP_ME_FOR_NOTEBOOKS_V4/DATA/Advertising.csv')
X = df.drop('sales',axis=1)
y=df['sales']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet

ElasticNet()
```

```python
baseElasticModel = ElasticNet()
paramGrid = {'alpha':[.1,1,5,10,50,100],'l1_ratio':[.1,.5,.7,.96,.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV

gridmodel = GridSearchCV(estimator=baseElasticModel,param_grid=paramGrid,scoring='neg_mean_squared_error',cv=5, verbose=2)
```

```python
gridmodel.fit(X_train,y_train)
```

```python
gridmodel.best_estimator_
```

```python
pd.DataFrame(gridmodel.cv_results_)
```

```python
my_preds = gridmodel.predict(X_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,my_preds)
```
