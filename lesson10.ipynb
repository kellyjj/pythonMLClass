```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv("08-Linear-Regression-Models/Advertising.csv")
df
```

```python
df['total_spent'] = df['TV']+df['radio']+df['newspaper']
df
```

```python
sns.scatterplot(data=df,x='total_spent',y='sales')
```

```python
sns.regplot(data=df,x='total_spent',y='sales')
```

```python
X = df['total_spent']
y = df['sales']
```

```python
#doig simple linear regression
# y= mx+b,  y = B1x+B0
betaCoEf =np.polyfit(X,y,deg=1)
betaCoEf
```

```python

potentialSpend = np.linspace(0,500,100)
betaCoEf[0]
#the coefficent returned by polyfit returns our coefficents in reverse order.  ie 2,1...
#so below,  betcoef[1 ] (since we only got 2 elements back would be B0)  betcoef[1] is our y-intercept
predictedSales = betaCoEf[0]*potentialSpend+betaCoEf[1]
```

```python
sns.scatterplot(data=df,x='total_spent',y='sales')
plt.plot(potentialSpend,predictedSales,color='red')
```

```python
spend = 200

predictedSales = betaCoEf[0]*spend+betaCoEf[1]
predictedSales
```

```python
newpoly =np.polyfit(X,y,deg=3)
newpoly

```

```python
potspend = np.linspace(0,500,100)
predsale = newpoly[0]*(potspend**3)+newpoly[1]*(potspend**2)+newpoly[2]*(potspend)+newpoly[3]
sns.scatterplot(data=df,x='total_spent',y='sales')
plt.plot(potspend,predsale,color='red')
```
